Understanding who is your customer is one of the most important tasks a successful business can do. This task helps businesses create, communicate, and deliver their offers by tailoring them to their customer's needs.

The aim of this project is to apply a clustering technique into the data to separate customers in a supermarket into segments. The underlying idea is that most likely this groups would have comparable purchase behaviors. In addition, these groups are most likely to respond to marketing messages in the same way.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage

import pickle

seed = 101096
```

```python
df_segmentation = pd.read_csv("Data/segmentation data.csv", index_col=0)
df_segmentation
```

**Standardizing data**

```python
scaler = StandardScaler()
array_segmentation_standard = scaler.fit_transform(df_segmentation.values)
array_segmentation_standard
```

```python
array_segmentation_standard.shape
```

**Hierarchical Clustering**

```python
hierarchical_clustering = linkage(array_segmentation_standard,
                                  method = 'ward')
```

```python
plt.figure(figsize = (12,9))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hierarchical_clustering,
           truncate_mode = 'level',
           p = 5)
plt.show()
```

Rule of thumb: We find the longest vertical line unintercepted by a horizontal line from the dendrogram. This is where we should make the cut.

**Princial Component Analysis**

```python
pca = PCA()
array_segmentation_standard_pca = pca.fit_transform(array_segmentation_standard)
array_segmentation_standard_pca
```

```python
array_segmentation_standard_pca.shape
```

All values of `.explained_variance_ratio_` add to 1

```python
pca.explained_variance_ratio_
```

```python
plt.plot(range(1, len(pca.explained_variance_ratio_)+1),
         pca.explained_variance_ratio_.cumsum(),
         marker='o')
plt.title('Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance')
plt.axhline(y=0.8, color='r');
```

Rule of thumb is to choose any number of components in which the cumulative variance is above 80%. In this case we can choose either 3 or 4. We will choose 3.

```python
pca_3 = PCA(n_components=3)
pca_3_scores = pca_3.fit_transform(array_segmentation_standard)
pca_3_scores
```

```python
pca_3_scores.shape
```

```python
pca_3.components_
```

```python
df_pca_3 = pd.DataFrame(data=pca_3.components_,
                        columns=df_segmentation.columns.values,
                        index=['Component 1', 'Component 2', 'Component 3'])
df_pca_3
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df_pca_3, annot=True, cmap='RdBu', vmin=-1, vmax=1)
plt.title('Correlation Heatmap Components');
```

```python
def display_components(df_pca, component_number):
    component_index = {1:'Component 1',
                       2:'Component 2',
                       3:'Component 3'}
    abs_order = np.abs(df_pca.loc[component_index[component_number], :]).sort_values(ascending=False).index.values
    component_values = df_pca.loc[component_index[component_number], abs_order]
    new_df = pd.DataFrame(component_values).reset_index()
    
    sns.barplot(data = new_df,
                x = component_index[component_number],
                y = 'index',
                palette = 'Blues_d')
    plt.xlabel('weights')
    plt.ylabel('features')
    plt.title(component_index[component_number])
```

Component 1 has a positive correlation with income, occupation, age, settlement size. This component relates to the career of a person. 

```python
display_components(df_pca_3 , 1)
```

Component 2 has a positive correlation with education, marital status, sex. This compoment relates to the individuals eduaction and lifestyle. 

```python
display_components(df_pca_3 , 2)
```

Component 3 has a positive correlation with age and negative correlation with marital status and occupation. This component relates to experience of a person. 

```python
display_components(df_pca_3 , 3)
```

**K-Means**

```python
clusters=[]
inertias=[]

for i in range(1,10):
    
    clusters.append(i)
    #'k-means++' is an algorithm that runs before the actual KMeans and finds the best starting points for the centroids
    model = KMeans(n_clusters=i, init='k-means++', random_state = seed)  
    model.fit(pca_3_scores)
    inertias.append(model.inertia_)
    
plt.plot(clusters, inertias, marker='o')
plt.xlabel('# of clusters')
plt.ylabel('inertia')
plt.title('Inertia vs # of clusters')
plt.axvline(x=4, color='r');
```

```python
model_4 = KMeans(n_clusters=4, init='k-means++', random_state = seed)
model_4.fit(pca_3_scores)
```

```python
component_dataframe = pd.DataFrame(pca_3_scores, columns=['Component 1', 'Component 2', 'Component 3'])
component_dataframe
```

```python
df_concat = pd.concat((df_segmentation.reset_index(drop=True), component_dataframe), axis=1)
df_concat.head()
```

```python
df_concat['cluster'] = model_4.labels_
df_clusters = df_concat.groupby('cluster').mean()
df_clusters
```

```python
len(model_4.labels_)
```

```python
df_clusters['Number of observations'] = df_concat.groupby('cluster').count()['Sex']
df_clusters['Proportion'] = df_clusters['Number of observations'] / df_clusters['Number of observations'].sum()
```

```python
df_clusters
```

* Cluster 0: Cluster 0 is both high in component 1 and 2 which means high in education, lifestyle and career. This cluster will be called "well-off".   
* Cluster 1: Cluster 1 is both low in component 1 and compoent 2 which means low in education, lifestyle and career. This cluster will be called "fewer-opportunities" since those people got the least opportunities.
* Cluster 2: Cluster 2 is low in component 1 (career) and high in component 2 (education and life-style). This cluster will be called "standard".
* Cluster 3: Cluster 3 is high in component 1 (career) and low in component 2 (education and life-style). This cluster will be called "career-focused".

```python
df_clusters = df_clusters.rename(index = {0:'well-off', 1:'fewer-opportunities', 2:'standard', 3:'career-focused'})
df_clusters
```

**3D Visualization**

```python
axis_1_cluster0 = df_concat[df_concat['cluster']==0]['Component 1']
axis_2_cluster0 = df_concat[df_concat['cluster']==0]['Component 2']
axis_3_cluster0 = df_concat[df_concat['cluster']==0]['Component 3']

axis_1_cluster1 = df_concat[df_concat['cluster']==1]['Component 1']
axis_2_cluster1 = df_concat[df_concat['cluster']==1]['Component 2']
axis_3_cluster1 = df_concat[df_concat['cluster']==1]['Component 3']

axis_1_cluster2 = df_concat[df_concat['cluster']==2]['Component 1']
axis_2_cluster2 = df_concat[df_concat['cluster']==2]['Component 2']
axis_3_cluster2 = df_concat[df_concat['cluster']==2]['Component 3']

axis_1_cluster3 = df_concat[df_concat['cluster']==3]['Component 1']
axis_2_cluster3 = df_concat[df_concat['cluster']==3]['Component 2']
axis_3_cluster3 = df_concat[df_concat['cluster']==3]['Component 3']

fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(axis_1_cluster0, axis_2_cluster0, axis_3_cluster0, color='red', label='well-off')
ax.scatter(axis_1_cluster1, axis_2_cluster1, axis_3_cluster1, color='blue', label='fewer-opportunities')
ax.scatter(axis_1_cluster2, axis_2_cluster2, axis_3_cluster2, color='black', label='standard')
ax.scatter(axis_1_cluster3, axis_2_cluster3, axis_3_cluster3, color='green', label='career-focused')

ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_zlabel('Component 3')
ax.set_title('Customer Segmentation Using \n PCA Components')
ax.legend(loc='best');
```
