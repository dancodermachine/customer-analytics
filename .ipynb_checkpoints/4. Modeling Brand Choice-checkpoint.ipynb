# Modeling Brand Choice

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression

import pickle
```

**Loading Models**

```python
#Importing models
df = pd.read_csv('final_data.csv')
df_incidence = df[df['Incidence']==1]
df_incidence.shape
```

```python
df_incidence
```

```python
Y = df_incidence['Brand']
x  = df_incidence[['Price_1', 'Price_2', 'Price_3', 'Price_4', 'Price_5']]
model_brand_choice = LogisticRegression(solver = 'sag', multi_class = 'multinomial')
model_brand_choice.fit(x, Y)
```

```python
model_brand_choice.coef_
```

Transpose Values

```python
bc_coef = pd.DataFrame(np.transpose(model_brand_choice.coef_))
bc_coef = bc_coef.rename(columns={0:'Coef_Brand_1',
                                  1:'Coef_Brand_2',
                                  2:'Coef_Brand_3',
                                  3:'Coef_Brand_4',
                                  4:'Coef_Brand_5'},
                         index={0:'Price_1',
                                1:'Price_2',
                                2:'Price_3',
                                3:'Price_4',
                                4:'Price_5'})
```

```python
bc_coef
```

### Price Elasticity of Brands

```python
name_list = ['Price_1', 'Price_2', 'Price_3', 'Price_4', 'Price_5']
means = {name:df_incidence[name].mean() for name in name_list}
price_range = np.arange(0.5, 3.5, 0.01)

def price_elasticity_brands(price_number, df_coef):
    """
    price_number should be between 1 and 5
    """
    price_number_str = str(price_number)
    df_own_brand = pd.DataFrame({'Price_' + price_number_str:price_range})
    for name in name_list:
        if df_own_brand.columns[0] != name:
            df_own_brand[name] = means[name]
            
    df_own_brand = df_own_brand[name_list]
            
    purchase_probability = model_brand_choice.predict_proba(df_own_brand)

    pp = purchase_probability[:,price_number-1]
    pp2 = purchase_probability[:][:,price_number-1]
        
    beta = df_coef.iloc[price_number-1, price_number-1]
        
    elasticity = beta * price_range * (1-pp)
            
    return elasticity
```

```python
brand1_elasticity = price_elasticity_brands(1, bc_coef)
brand2_elasticity = price_elasticity_brands(2, bc_coef) 
brand3_elasticity = price_elasticity_brands(3, bc_coef) 
brand4_elasticity = price_elasticity_brands(4, bc_coef) 
brand5_elasticity = price_elasticity_brands(5, bc_coef) 
```

```python
brands_elasticities =  pd.DataFrame({'price_range':price_range,
                                     'brand1_elasticity':brand1_elasticity,
                                     'brand2_elasticity':brand2_elasticity,
                                     'brand3_elasticity':brand3_elasticity, 
                                     'brand4_elasticity':brand4_elasticity, 
                                     'brand5_elasticity':brand5_elasticity})
```

```python
brands_elasticities
```

```python
plt.figure(figsize=(8,8))
plt.plot('price_range', 'brand1_elasticity', data=brands_elasticities)
plt.plot('price_range', 'brand2_elasticity', data=brands_elasticities)
plt.plot('price_range', 'brand3_elasticity', data=brands_elasticities)
plt.plot('price_range', 'brand4_elasticity', data=brands_elasticities)
plt.plot('price_range', 'brand5_elasticity', data=brands_elasticities)
plt.xlabel('price')
plt.ylabel('elasticity')
plt.title('Own Price Elasticity of Purchase Probability for different Brands')
plt.legend();
```
