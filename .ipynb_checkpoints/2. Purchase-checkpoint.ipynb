# Purchase

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

import pickle
```

```python
df_chocolate = pd.read_csv('Data/purchase data.csv')
df_chocolate.head()
```

```python
df_chocolate.info()
```

## Segmentation

```python
scaler = pickle.load(open('Segmentation Models/scaler.pickle', 'rb'))
pca = pickle.load(open('Segmentation Models/pca.pickle', 'rb'))
kmeans_pca = pickle.load(open('Segmentation Models/kmeans_pca.pickle', 'rb'))
```

**Standardization**

```python
df_features = df_chocolate[df_chocolate.columns.values[-7:]]
df_features
```

```python
array_chocolate_stan = scaler.transform(df_features)
array_chocolate_stan
```

```python
array_chocolate_stan.shape
```

**PCA**

```python
array_chocolate_stan_pca = pca.transform(array_chocolate_stan)
array_chocolate_stan_pca
```

```python
array_chocolate_stan_pca.shape
```

**K-means PCA**

```python
kmeans_values = kmeans_pca.predict(array_chocolate_stan_pca)
kmeans_values
```

```python
df_chocolate_predict = df_chocolate.copy()
df_chocolate_predict['Cluster'] = kmeans_values
```

```python
df_chocolate_predict
```

## Descriptive Analytics

**Consumers**

```python
df_chocolate_predict.head()
```

```python
#Number of purchases per individual
num_visits = df_chocolate_predict[['ID', 'Incidence']].groupby('ID', as_index=False).count()
purchases = df_chocolate_predict[['ID', 'Incidence']].groupby('ID', as_index=False).sum()
cluster = df_chocolate_predict[['ID', 'Cluster']].groupby('ID', as_index=False).mean()
```

```python
df_customer = num_visits.merge(purchases, on='ID')
df_customer.rename(columns={'Incidence_x':'N_Visits', 'Incidence_y':'N_Purchases'}, inplace=True)
df_customer['Average_purchases_pervisit'] = df_customer['N_Purchases']/df_customer['N_Visits']
df_customer = df_customer.merge(cluster, on='ID')
```

```python
df_customer
```

**Segment**

```python
total = df_customer[['ID', 'Cluster']].groupby('Cluster').count().sum()[0]
segment = df_customer[['ID', 'Cluster']].groupby('Cluster').count()/total
segment_prop = segment.rename(columns={'ID':'Prop'})
segment_prop
```

```python
plt.figure(figsize=(9,6))
plt.pie(segment_prop['Prop'],
       labels=['Well-off','Fewer-opportunities','Standard','Career-focused'],
       autopct='%1.1f%%')
plt.title('Segment Proportions');
```

**Purchase Ocasion and Purchase Incidance**

We will find a quantitative measure to describe the purchase behaivours by segments. How often do they visit stores? How often do they buy chocolate bars? How often they buy relative to their visits to store?

```python
segment_mean = df_customer.groupby('Cluster').mean().drop(columns=['ID']).reset_index()
segment_std = df_customer.groupby('Cluster').std().drop(columns=['ID']).reset_index()
data_segment = segment_mean.merge(segment_std, on='Cluster', suffixes=('_mean', '_std'))
data_segment
```

```python
plt.figure(figsize=(9,6))
sns.barplot(data=df_customer, x='Cluster', y='N_Visits', ci='sd')
plt.ylabel('Number of Stores Visits')
plt.xlabel('Cluster')
plt.title('Average Number of Store Visits by Cluster')
plt.xticks([0,1,2,3], ['Well-off','Fewer-opportunities','Standard','Career-focused']);
```

```python
plt.figure(figsize=(9,6))
sns.barplot(data=df_customer, x='Cluster', y='N_Purchases', ci='sd')
plt.ylabel('Purchase Incidences')
plt.xlabel('Cluster')
plt.title('Number of Purchases by Segment')
plt.xticks([0,1,2,3], ['Well-off','Fewer-opportunities','Standard','Career-focused']);
```

```python
plt.figure(figsize=(9,6))
sns.barplot(data=df_customer, x='Cluster', y='Average_purchases_pervisit', ci='sd')
plt.ylabel('Average Purchase per Visit')
plt.xlabel('Cluster')
plt.title('Average Purchase per Visit by Segment')
plt.xticks([0,1,2,3], ['Well-off','Fewer-opportunities','Standard','Career-focused']);
```

**Brand Choice**

```python
df_chocolate_incidence=df_chocolate_predict[df_chocolate_predict['Incidence']==1]
brand_dummies = pd.get_dummies(df_chocolate_incidence['Brand'], prefix='Brand', prefix_sep='_')
```

```python
brand_dummies['Cluster']=df_chocolate_incidence['Cluster']
brand_dummies['ID']=df_chocolate_incidence['ID']
brand_dummies.reset_index(drop=True)
```

```python
average_brand_percustomer = brand_dummies.groupby('ID').mean()
average_brand_perccluster = average_brand_percustomer.groupby('Cluster').mean()
average_brand_perccluster
```

```python
sns.heatmap(data=average_brand_perccluster,
            vmin=0,
            vmax=1,
            cmap='PuBu',
            annot=True);
plt.title('Average Brand Choice per Cluster')
plt.yticks([0,1,2,3], ['Well-off','Fewer-opportunities','Standard','Career-focused'],
           rotation=30);
#Brand 1 is the cheapest and Brand 5 is the most expensive
```

**Revenue**

```python
brand1 = df_chocolate_predict[df_chocolate_predict['Brand']==1].copy()
brand1['Revenue_brand1'] = brand1['Quantity']*brand1['Price_1']
brand1_revenue = brand1[['Cluster','Revenue_brand1']].groupby('Cluster', as_index=False).sum()

brand2 = df_chocolate_predict[df_chocolate_predict['Brand']==2].copy()
brand2['Revenue_brand2'] = brand2['Quantity']*brand2['Price_2']
brand2_revenue = brand2[['Cluster','Revenue_brand2']].groupby('Cluster', as_index=False).sum()

brand3 = df_chocolate_predict[df_chocolate_predict['Brand']==3].copy()
brand3['Revenue_brand3'] = brand3['Quantity']*brand3['Price_3']
brand3_revenue = brand3[['Cluster','Revenue_brand3']].groupby('Cluster', as_index=False).sum()

brand4 = df_chocolate_predict[df_chocolate_predict['Brand']==4].copy()
brand4['Revenue_brand4'] = brand4['Quantity']*brand4['Price_4']
brand4_revenue = brand4[['Cluster','Revenue_brand4']].groupby('Cluster', as_index=False).sum()

brand5 = df_chocolate_predict[df_chocolate_predict['Brand']==5].copy()
brand5['Revenue_brand5'] = brand5['Quantity']*brand5['Price_5']
brand5_revenue = brand5[['Cluster','Revenue_brand5']].groupby('Cluster', as_index=False).sum()
```

```python
a = pd.merge(brand1_revenue, brand2_revenue, on='Cluster')
b = pd.merge(brand3_revenue, brand4_revenue, on='Cluster')
c = pd.merge(b, brand5_revenue, on='Cluster')
revenues = a.merge(c, on='Cluster')
revenues['Total'] = (revenues['Revenue_brand1'] + revenues['Revenue_brand2'] + revenues['Revenue_brand3'] +
                     revenues['Revenue_brand4'] + revenues['Revenue_brand5'])
revenues['Cluster Proportion'] = segment_prop['Prop']
revenues['Cluster'] = ['Well-off','Fewer-opportunities','Standard','Career-focused']
```
